# Preprocessing 
EDA is done in a separate notebook.

```python
df = pd.read_csv("ingredients.csv", sep = ";", header=1)
df.dtypes
```

```python
df["Descrip"].str.lower().str.strip().duplicated().sum()
```

```python
df = df.drop_duplicates()
excl = ["Descrip"]
cols_conv = df.columns.difference(excl)
df[cols_conv] = df[cols_conv].apply(
    lambda col: pd.to_numeric(col, errors="coerce")
)
```
